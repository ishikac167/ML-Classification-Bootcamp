# STEP #1: PROBLEM STATEMENT

You work as a data scientist at a major bank in NYC and you have been tasked to develop a model that can predict whether a customer is able to retire or not based on his/her features. Features are his/her age and net 401K savings (retirement savings in the U.S.). You though that Support Vector Machines can be a great candidate to solve the problem. 

# STEP #2: IMPORTING DATA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
bank = pd.read_csv('Bank_Customer_retirement.csv')
```

```python
bank.keys()
```

```python
bank.shape
```

```python
bank.head()
```

```python
bank.tail()
```

# STEP #3: VISUALIZING THE DATA

```python
sns.pairplot(bank, hue='Retire', vars=['Customer ID', 'Age', '401K Savings'])
```

```python
sns.countplot(bank['Retire'], label='Retirement')
```

# STEP #4: MODEL TRAINING 

```python
bank = bank.drop(['Customer ID'], axis=1)
```

```python

# Let's drop the target label coloumns
X = bank.drop(['Retire'], axis=1)
```

```python
X
```

```python
y = bank['Retire']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train, y_train)
```

# STEP #5: EVALUATING THE MODEL

```python
y_predict = svc_model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

# STEP #6: IMPROVING THE MODEL

```python
min_train = X_train.min()
min_train
```

```python
range_train = (X_train-min_train).max()
range_train
```

```python
X_train_scaled = (X_train-min_train)/range_train
```
